```python
x = []
y = []
for i in range(-50,100,5):
    x.append(i)
    y.append(i)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,s=100)
```

```python
plt.scatter(x,y,s=30,c='green')
```

```python
plt.scatter(x,y,s=30,c='green',marker='+')
```

```python
x = []
y = []
y2 = []
y3 = []

for i in range(-50,100,5):
    x.append(i)
    y.append(i)
    y2.append(i*2)
    y3.append(i+50)
```

```python
plt.scatter(x,y,marker='o')
plt.scatter(x,y2, c='gray')
plt.scatter(x,y3, c='green',marker='+')
```
