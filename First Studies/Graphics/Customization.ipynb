```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[]
y=[]
y2=[]

for num in range(-200,201,10):
    x.append(num)
    y.append(num)
    y2.append(2*num)
```

```python
plt.plot(x,y,c='red',ls='-',lw=2,marker='o',ms=10,fillstyle='top',mfc='blue',
        label = 'Primeiro')
plt.plot(x,y2,c='green',
        label = 'Segundo')

# Adicionando label nos eixos
plt.xlabel('Eixo X', color='red',fontsize=15,horizontalalignment='center',verticalalignment='center')
plt.ylabel('Eixo Y', color='blue',fontsize=15,horizontalalignment='center',verticalalignment='center')
plt.title(r'Título $\frac{5}{4}$',color='gray',fontsize=25,horizontalalignment='center')
plt.legend(loc='upper left')

# Customizando ticks
eixoX = ['Jan','Abr', 'Jul']
eixoY = ['Fraco','Regular','Neutro','Bom','Excelente']
plt.xticks([-200,0,200],eixoX, rotation=30) # formato é ([de], para)
plt.yticks([-400,-200,0,200,400],eixoY)
```
