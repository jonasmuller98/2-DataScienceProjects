```python
import numpy as np
import pandas as pd
```

```python
drinks = pd.read_csv('drinks.csv')
```

```python
def consumo(a):
    if a <= 75:
        return 'low'
    if a >= 151:
        return 'high'
    else:
        return 'middle'
```

```python
# Apply aplica função ao longo do dataframe de acordo com a coluna indicada
drinks['consumo'] = drinks['beer_servings'].apply(consumo)
```
